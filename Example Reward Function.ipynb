```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def test_function(x):
    fx = np.sin(np.pi/2.0 - 5.0 * x) + np.sin(0.5 * x) + 2.0 * np.sin(2 * x)
    return fx
```

```python
x = np.array(np.linspace(-4, 5, num=250))
y = np.array([test_function(i) for i in x])
```

```python
plt.plot(x,y)
plt.xlabel('Action')
plt.ylabel('Reward')
plt.title('Example Continous Reward Function')
```
